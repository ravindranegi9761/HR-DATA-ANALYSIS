# deal with outliers in column

```python
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
```

```python
numerical_columns=df1.select_dtypes(include=['number']).columns.tolist()
numerical_columns
```

```python
fig,axes=plt.subplots(nrows=len(numerical_columns),ncols=1,figsize=(10,6*len(numerical_columns)))
for i,col in enumerate(numerical_columns):
    sns.boxplot(x=df1[col],ax=axes[i])
plt.tight_layout()
plt.show()
```

```python
for i in numerical_columns:
    z_score=stats.zscore(df1[col])
    thresold=3
    outliers_indices=(z_score>thresold) | (z_score<-thresold)
df1[outliers_indices==True]   
```

```python
df2=df1[~outliers_indices]
df2
```

```python
df2.reset_index(drop=True, inplace=True)
```
